```python
#importing the libararies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Setting the size of the plots (10 inches width and 10 inches height)
plt.rcParams['figure.figsize']=(10.0,10.0)
#loading the data
data=pd.read_csv('details.csv')
#Removing rows and colums with missing values
data=data.dropna()
data=data.dropna(axis=1)
data.head()
```

```python
#Plotting (Visualization using Scatter plot)
plt.scatter(data.Hours, data.Scores)
plt.xlabel("Hours Studied")
plt.ylabel("Scores")
plt.show()
```

```python
#Gradient Descent
def gradient(m, b, L):
    n = len(data)
    x = data.Hours
    y = data.Scores
    
    y_pred = m * x + b
    error = y - y_pred
    
    m_gradient = (-2/n) * sum(x * error)
    b_gradient = (-2/n) * sum(error)
    m = m - L * m_gradient
    b = b - L * b_gradient
    return m, b
```

```python
#Estimating the value of m and b
m=0
b=0
L=0.001
epochs=100
for i in range(epochs):
    m,b=gradient(m,b,L)

m_value = m
b_value = b
```

```python
#Plotting linear regression
plt.scatter(data.Hours, data.Scores)
plt.plot(range(11), [m_value * x + b_value for x in range(11)], color='red')
plt.xlabel("Hours Studied")
plt.ylabel("Scores")
plt.show()
```
